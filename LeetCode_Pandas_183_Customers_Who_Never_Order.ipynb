## 183. Customers Who Never Order

Table: `Customers`

| Column Name | Type    |
|-------------|---------|
| id          | int     |
| name        | varchar |

* id is the primary key (column with unique values) for this table.
Each row of this table indicates the ID and name of a customer.


Table: Orders

| Column Name | Type |
|-------------|------|
| id          | int  |
| customerId  | int  |

* id is the primary key (column with unique values) for this table.
customerId is a foreign key (reference columns) of the ID from the Customers table.
* Each row of this table indicates the ID of an order and the ID of the customer who ordered it.


<b>Write a solution to find all customers who never order anything.

Return the result table in any order.

The result format is in the following example.</b>

Example 1:

Input:
Customers table:

| id | name  |
|----|-------|
| 1  | Joe   |
| 2  | Henry |
| 3  | Sam   |
| 4  | Max   |


Orders table:

| id | customerId |
|----|------------|
| 1  | 3          |
| 2  | 1          |

Output:

| Customers |
|-----------|
| Henry     |
| Max       |


```python
import pandas as pd

# Creating customer data frame
df = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['Joe', 'Henry', 'Sam', 'Max']
  })
df
```

```python
# Creating orders data frame
df2 = pd.DataFrame({
    'id': [1, 2],
    'customerId': [3, 1]
  })
df2
```

```python
# Customer who ordered
df[df['id'].isin(df2['customerId'])]
```

```python
# Customer who never ordered
df[~df['id'].isin(df2['customerId'])]['name']
```
